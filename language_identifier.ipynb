```python
language_list = list(language_dict.values())[:100]
```

```python
import wikipedia
data = []
labels = []
for i, lan in enumerate(tqdm(language_list)):
    wikipedia.set_lang(lan)
    lan_data = []
    for topic in topic_list:
        try:
            lan_data.append(wikipedia.summary(topic))
        except wikipedia.exceptions.DisambiguationError as e:
            lan_data.append(wikipedia.summary(e.options[0]))
        except wikipedia.exceptions.PageError as e:
            continue
    data.extend(lan_data)
    labels.extend([i] * len(lan_data))
```
